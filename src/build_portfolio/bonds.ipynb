```python
import polars as pl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

figsize = (1.618*6, 6)

path = '../../data/NEXTFUNDS/2511_country_ratio.xlsx'
df = pd.read_excel(path, sheet_name='region')
nf_foreign_bonds = pl.DataFrame(df)
path = '../../data/NEXTFUNDS/2519_country_ratio.xlsx'
df = pd.read_excel(path, sheet_name='region')
nf_emerging_bonds = pl.DataFrame(df)
```

```python
nf_foreign_bonds
```

```python
df = pl.DataFrame({
    'region': ['north_america', 'south_america', 'emea', 'asia_pasific'],
    'resio': [nf_foreign_bonds['ratio'][0], nf_foreign_bonds['ratio'][1], nf_foreign_bonds['ratio'][2], nf_foreign_bonds['ratio'][3]]
})
df
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.set_title('NEXT FUNDS FTSE Developed Bond ETF')
ax.pie(nf_foreign_bonds['ratio'], labels=nf_foreign_bonds['region'], autopct='%1.1f%%', startangle=90, counterclock=False);
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.set_title('NEXT FUNDS Emerging Bond ETF')
ax.pie(nf_emerging_bonds['ratio'], labels=nf_emerging_bonds['region'], autopct='%1.1f%%', startangle=90, counterclock=False);
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.set_title('NEXT FUNDS Emerging Bond ETF')
ax.pie(df['ratio'], labels=df['region'], autopct='%1.1f%%', startangle=90, counterclock=False);
```
