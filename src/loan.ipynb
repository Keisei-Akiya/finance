# 借金返済の金利

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

path = '../data/loan_rate.xlsx'
df = pd.read_excel(path)
df.set_index('month', inplace=True)

figsize = (1.618*6, 6)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.plot(df.query('year > 0').index, df.query('year > 0')['f_rate'], label='固定')
ax.plot(df.query('year > 0').index, df.query('year > 0')['r_rate'], label='変動')
# ax.axvline(13, ls='--', color='gray')
# ax.axvline(25, ls='--', color='gray')
# ax.axvline(37, ls='--', color='gray')
ax.set_title('金利の推移 (月次)')
ax.set_xlabel('月')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=figsize)
ax.plot(df.index, df['f_loan_h'], label='固定')
ax.plot(df.index, df['r_loan_h'], label='変動')
ax.axhline(y=0, color='black', linewidth=0.5, label='完済')
ax.set_title('固定金利 (2%) vs 変動金利 (1, 2, 3, 4%)')
ax.set_xlabel('月')
ax.set_ylabel('残高')
ax.legend()
```
