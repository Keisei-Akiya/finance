```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

from get_geometric_mean_annual_return import get_geometric_mean_annual_return
```

```python
ticker = '^TOPX'
topix_data = yf.download(ticker, start='2010-01-01', end='2024-09-20')
topix_data.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(topix_data['Close'], label='TOPIX')
plt.show()
```

```python
geometric_mean_annual_return_topix = get_geometric_mean_annual_return(topix_data)
print(f'TOPIX 年次幾何平均リターン: {geometric_mean_annual_return_topix:.2f}%')
```

```python
# 日次リターンの標準偏差を計算する
daily_volatility = topix_data['Return'].std()

# 年次リスク（標準偏差）を計算する
annual_volatility = daily_volatility * np.sqrt(252)  # 252は1年の取引日数の平均

print(f"TOPIXの年次リスク（標準偏差）: {annual_volatility:.6f}")
```
