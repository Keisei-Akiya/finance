# Pie chart

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np
```

```python
asset_balance = {
    "株式": 30,
    "債券": 50,
    "不動産": 10,
    "コモディティ": 10
}

keys = asset_balance.keys()
values = asset_balance.values()

fig, ax = plt.subplots(figsize=(1.618*7, 7))
ax.set_title("Asset Allocation", fontsize=20)
ax.pie(values, labels=keys, autopct="%1.1f%%", startangle=90, counterclock=False)
ax.axis("equal")
plt.show()
```
