```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

from get_geometric_mean_annual_return import get_geometric_mean_annual_return
```

```python
ticker = '^TOPX'
topix_data = yf.download(ticker, start='2010-01-01', end='2024-09-20')
topix_data.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(topix_data['Close'], label='TOPIX')
plt.show()
```

```python
geometric_mean_annual_return_topix = get_geometric_mean_annual_return(topix_data)
print(f'TOPIX : {geometric_mean_annual_return_topix:.2f}%')
```

```python
# 
daily_volatility = topix_data['Return'].std()

# 
annual_volatility = daily_volatility * np.sqrt(252)  # 2521

print(f"TOPIX: {annual_volatility:.6f}")
```
